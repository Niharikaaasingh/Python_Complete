### Q2:- Write a program that will convert celsius value to fahrenheit.

````python
# Write your code here

# Program to convert Celsius to Fahrenheit

# Take input from user
celsius = float(input("Enter temperature in Celsius: "))

# Conversion formula
fahrenheit = (celsius * 9/5) + 32

# Display result
print(f"{celsius}°C is equal to {fahrenheit}°F")
````

### Q4:- Write a program to find the euclidean distance between two coordinates.Take both the coordinates from the user as input.

````python
# Write your code here
import math
x1= float(input("enter value1"))
x2= float(input("enter value2"))
y1= float(input("enter value3"))
y2= float(input("enter value4"))

distance = math.sqrt((x2-x1)**2 + (y2-y1)**2)

print(f"The euclidean distance between({x1},{y1}) and ({x2},{y2} is {distance})")
````

### Q6:- Write a program that will tell the number of dogs and chicken are there when the user will provide the value of total heads and legs.

For example:
Input:
heads -> 4

<br>

dogs -> 2
chicken -> 2





````python
# Write your code here
# Inputs
H = int(input("Enter total number of heads: "))
L = int(input("Enter total number of legs: "))

# Direct calculation
D = (L - 2*H) // 2
C = H - D

# Output
print("Number of chickens =", C)
print("Number of dogs =", D)

````
