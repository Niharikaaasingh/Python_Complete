#List Comprehension

```python
L = [i for i in range(1,11)]
print(L)
```

```python
#scalar multiplication on a vector
v =[2,3,4]
s = -3
[s*i  for i in v]
```

```python
# Add squares
L = [2,3,4,5,6,7]

[i**2 for i in L]
```

```python
#print all numbers divisible by 5 in the range of 1 to 50
[i for i in range(1,51) if i%5 == 0]
```

```python
#find languages which start with letter p
languages = ['java','python','R','C','javascript']

[language for language in languages  if language.startswith('p')]

```

```python
#nested if with list comprehension
basket =['apple','banana','cherry','strawberry']
my_fruits =[ 'apple','kiwi','grapes','banana']

[fruit for fruit in my_fruits if fruit in basket if fruit.startswith('a') ]
```

```python
#print a (3,3) matrix usinglist comprehension -> Nested list comprehension
[[i*j for i in range(1,4)] for j in (1,4)]
```

```python
#contains prioducts -> list comprehension on 2 lists together
L1 = [1,2,3,4]
L2 = [5,6,7,8]

[i*j for i in L1 for j in L2]
```

# 2 Ways to Traverse a List
* itemwise
* indexwise

```python
#itemwise
L = [1,2,3,4]

for i in L:
  print(i)
```

```python
#indexwise
L = [1,2,3,4]

for i in range(0,len(L)):
  print(L[i])
```

#ZIP
 the zip() function return a zip object which is an iterator of tuples where the first item in each passed iterator is paired together, and the second item in each passed iterator are paired together .

 if the passed iterators have different lengths ,the iterator with the least items decides the length of the new iterator,

```python
#wap  to add items of 3 lists indexwise
L1=[1,2,3,4]
L2 =[-1,-2,-3,-4]

list(zip(L1,L2))

[i+j for i,j in zip(L1,L2) ]
```
