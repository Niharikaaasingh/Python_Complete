# OPERATORS

```python
#Arithmetic operator
print(5+6)
print(5*6)
print(5/6)
print(5//6)
print(5%2)
print(5**2)
```

```python
#relational operator
print(4>5)
print(5<9)
print(4<=4)
print(5>=5)
print(5!=3)
print(9==9)
```

```python
#Logical OPerator
print(1 and 0)

print(1 or 0)

print(not 1)
```

```python
#bitwise operator  operate on binary values

#bitwise and
print(2 & 3)

#btwise or
print(2 | 3)

#bitwise xor
print(2 ^ 3)  #diff 1 same 0

#bitwise notp
print(~3)

#bitwise left operator
print(4 >> 2)

#bitwise right operator
print(5 << 2)
```

```python
# Assignment operator
a = 2
a %= 2
print(a)
```

```python
#membership operator

#in/not  in

print('D' not in 'Delhi')

print(1 in [2,4,5,6,7,1])
```

```python
#menu driven calculator
fnum = int(input('enter the 1st number: '))
snum = int(input('enter 2nd num: '))

op = input('enter the operations')

if op == '+':
  print(fnum + snum)
elif op == '-':
   print(fnum + snum)
elif op == '*':
   print(fnum * snum)
elif op == '/':
   print(fnum / snum)
elif op == '**':
   print(fnum ** snum)
else:
   print(fnum % snum)
```

```python
# While loop example -> program to print the table
# Program -> Sum of all digits of a given number
# Program -> keep accepting numbers from users till he/she enters a 0 and then find the avg
```

```python
number = int(input('enter the number : '))
i = 1
while i < 11:
  print(number,'*',i,'=',number * i )
  i += 1

```

```python
#while loop with else
x = 2

while x < 10:
  print(x)
  x += 1
else:
  print('limit crossed')

```

```python
#guesssing game

#generate random integer beteen 1 and 100
import random
jackpot = random.randint(1,100)

guess = int(input("guess karo"))
counter = 1

while guess != jackpot:
  if guess < jackpot:
    print("guess higher")
  else:
    print("guess lower")

  guess = int(input('guess karo'))
  counter += 1

else:
  print("correct guess")
  print("attempt",counter)







```

### Program - The current population of a town is 10000. The population of the town is increasing at the rate of 10% per year. You have to write a program to find out the population at the end of each of the last 10 years.

```python
#for loop examples
curr_pop = 10000

for i in range(10,0,-1):
  print(i,curr_pop)
  curr_pop = curr_pop - 0.1*curr_pop

```
